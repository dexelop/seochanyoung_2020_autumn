# 앤스콤 데이터 집합 불러오기

```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

# matplotlib 라이브러리로 간단한 그래프 그리기

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(anscombe)
df.head()
```

```python
df_dataset1 = df[df["dataset"]=="I"]
# df_dataset1.plot()
plt.plot(df_dataset1["x"], df_dataset1['y'],'o')
```

```python
df_dataset1 = df[df["dataset"]=="I"]
df_dataset2 = df[df["dataset"]=="II"]
df_dataset3 = df[df["dataset"]=="III"]
df_dataset4 = df[df["dataset"]=="IV"]
```

```python
fig = plt.figure()
```

```python
axis1 = fig.add_subplot(2,2,1)
axis2 = fig.add_subplot(2,2,2)
axis3 = fig.add_subplot(2,2,3)
axis4 = fig.add_subplot(2,2,4)
```

```python
axis1.plot(df_dataset1['x'], df_dataset1['y'],'o')
axis2.plot(df_dataset2['x'], df_dataset2['y'],'o')
axis3.plot(df_dataset3['x'], df_dataset3['y'],'o')
axis4.plot(df_dataset4['x'], df_dataset4['y'],'o')
```

# 한 번에 4개의 그래프 그리기

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = plt.figure()
axis1 = fig.add_subplot(1,1,1)
```

```python
axis1.hist(tips['total_bill'], bins=10)
axis1.set_title('Histogram of Total Bill')
axis1.set_xlabel('Frequency')
axis1.set_ylabel('Total Bill')
```

# 단변량 그래프 그리기 - 히스토그램

```python
scatter_plot = plt.figure()
axis1 = scatter_plot.add_subplot(1,1,1)
axis1.scatter(tips['total_bill'], tips['tip'])
axis1.set_title('Scatter Plot of Total Bill vs tip')
axis1.set_xlabel('Total Bill')
axis1.set_ylabel('Tip')
```
